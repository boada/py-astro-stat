##Gaussian Distribution##

```python
# Let's start with the distribution we've likely seen the most, the Gaussian distribution

# Import the stats module from scipy
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np

# Generate an instance of a normal distribution
# with a mean = 0 and stdev = 1
dist = stats.norm(0, 1)

# So that we can see what it looks like, let's make an x-axis using
# the percent point function of dist
x = np.linspace(dist.ppf(0.01), dist.ppf(0.99), num = 100)

# And display it for our sanity, but let's have the plot
# happen inline by using the ipython "magic" command
# %matplotlib inline
%matplotlib inline
plt.plot(x, dist.pdf(x), 'k-')
plt.xlabel('$x$')
plt.ylabel('$p(x|\mu, \sigma)$')
plt.show()
```

```python
# Now randomly sample the distribution 100 times
r100 = dist.rvs(size=100)

# And 1000 times
r1000 = dist.rvs(size=1000)

# And compare it to our curve
plt.hist(r100, normed=True, histtype='stepfilled', alpha=0.2, label='100 points')
plt.hist(r1000, normed=True, histtype='step', label='1000 points')
plt.plot(x, dist.pdf(x), 'k-', label='norm pdf')
plt.legend(loc='best', frameon=False)
plt.show()
```

##Uniform distribution & Monty Hall##

```python
# Now fun with a uniform distribution!
# I'm going to try to simulate the results of the Monty
# Hall problem (see example in sect. 3.1.3 of text)

# Start by generating a uniform distribution with its left
# edge at 1 and of width 3
dist = stats.uniform(1, 3)

# Plot it for our sanity
x = np.linspace(0, 5, num = 300)
plt.plot(x,dist.pdf(x))
plt.xlim([0,5])
```

```python
# With sanity intact, let's see if we can't do a Monty Hall simulation

def monty_hall(iterations):

    import random 

    # Sample the distribution 100 times, one sampling
    # represents my initial choice for door (rv_me), while
    # the other is the door the car is behind (rv_monty)
    rv_me = dist.rvs(iterations)
    rv_monty = dist.rvs(iterations)

    # Let's set up counters and doors
    doors = np.array((1,2,3))

    win_car_stay = 0
    win_goat_stay = 0

    win_car_change = 0
    win_goat_change = 0

    for i in range(iterations):
    
        # Let's have Monty choose a car
        car = int(rv_monty[i])
    
        # Figure out which door we've chosen
        chosen_door = int(rv_me[i])

        # Now play through the game. If I've selected the car, Monty will now
        # show me one of the other two doors, but since I'm not changing
        # my answer it doesn't matter what door he shows me. So, in the not 
        # changing my answer case, I just need to see if car == chosen_door
        if (car == chosen_door):
            win_car_stay += 1
        else:
            win_goat_stay += 1
            
        # In the case of changing our door, we need Monty to show us
        # a non-car door that we did not choose, or just get a goat
        # if we were (un)fortunate enough to choose the car initially
        if (car == chosen_door):
            win_goat_change += 1
        else:
            if car == 1:
                goats = np.array((2,3))
            if car == 2:
                goats = np.array((1,3))
            if car == 3:
                goats = np.array((1,2))
            
            # Now show the door with a goat I didn't choose
            door_show = goats[np.where(goats != chosen_door)]
            
            # And now pick the door that is not the door just opened
            # nor the one I'd already chosen
            new_chosen_door = doors[np.where((doors != door_show) & (doors != chosen_door))[0]][0]
                                                                                                    
            if (car == new_chosen_door):
                win_car_change += 1
            else:
                win_goat_change += 1
        
    # And print results
    print "Staying the decision:"
    print "     Success =", win_car_stay/float(iterations), \
          "| Failure =", win_goat_stay/float(iterations)

    print "Changing doors:"
    print "     Success =", win_car_change/float(iterations), \
          "| Failure =", win_goat_change/float(iterations)
```

```python
# With that function written, let's run it for 50 iterations
monty_hall(50)
```

```python
# And a 500 iterations
monty_hall(500)
```

```python
# And 5000 iterations
monty_hall(5000)
```

###Bayes' Rule###

Assuming two continuous random variables are not independent, it follows from the first eqn of the 2nd slide:
$$p(x,y) = p(x|y)p(y) = p(y|x)p(x) \,\, (1)$$

And the *marginal probability function* $p(x)$ is defined as 
$$p(x) = \int p(x,y) dy$$

Combining the above two equations
$$p(x) = \int p(x|y) p(y) dy \,\, (2)$$

Combining (1) and (2), we get ***Baye's rule***:

$$p(y|x) = \frac{p(x|y)p(y)}{p(x)} = \frac{p(x|y) p(y)}{\int p(x|y) p(y) dy}$$

or as a sum and to be used below

$$p(y_j|x) = \frac{p(x|y_j)p(y_j)}{p(x)} = \frac{p(x|y_j) p(y_j)}{\sum_{j=1}^N p(x|y_j) p(y_j)}$$

The textbook authors would like you to know that Bayes' rule relates conditional and marginal probabilities to each other and forms the foundation of Bayesian statistics.  Bayesian statistics has proved, until recently, to be controversial.  However, that is not because of this rule, but because of the rule's application.  We get to learn more about Bayesian stats in chapter 5.

###Now, Monty###

$H_i$ is the hypothesis that the car is behind the $i$th door.  For $N$ boxes, 
$$p(H_i|I) = 1/N \,\, (3)$$  

is the prior probability of selecting the car given background information $I$ (ie: one door has a car).  

Now, assume we choose the first door initially.  The "data" we're given that a goat is behind $N-2$ doors is represented $d_k$, (one door being the first door, which we chose and the $k$th being the other remaining closed door).  The probability we've correctly chosen the correct door with the car is

$$p(H_1|d_k, I) = \frac{p(d_k|H_1, I) p(H_1|I)}{p(d_k|I)} \,\,(4)$$

Additionally, the probability that the $k$th door remains closed because $H_1$ is true is $p(d_k|H_1,I) = 1/(N-1)$, as there are $N-1$ doors to choose from for goat-revealing purposes.

Now, using the law of total probability, the denominator becomes
$$p(d_k|I) = \sum_{i=1}^N p(d_k|H_i, I) p(H_i|I)$$

Additionally, the probability that the $k$th box does indeed stay shut (ie: the prize is behind the $i$th door) is
$$p(d_k|H_i, I) = \begin{cases}
1 \, \mathrm{for} \, k = i \\
0 \, \mathrm{otherwise}
\end{cases} \,\, (5)$$
except for when $i = 1$, which we discussed above.

The sum then becomes
$$p(d_k|I) = p(d_k|H_1,I)p(H_1|I) + p(d_k|H_k, I)p(H_k|I)$$

Which we can simplify to
$$p(d_k|I) = \frac{1}{N(N-1)} + \frac{1}{N} = \frac{1}{N-1} \,\, (6)$$

In the instance that we change our door selection from the 1st door to the $k$th door after being shown what's behind $N-2$ doors, we can rewrite eqn. (4) using eqns. (3), (5), and (6) to show
$$p(H_k|d_k, I) = \frac{1 \times (1/N)} {1/(N-1)} = \frac{N-1}{N}$$

In the case of $N=3$, as Monty liked to do, that means by switching from the initially selected 1st door to the only other remaining unopened $k$th door after the goat reveal (wordy...), our probability of getting a car is 2/3.

##Estimate of Bivariate Gaussian Distribution from Data##

###Pulled from section 3.5.3###

```python
# Let's try to recreate figure 3.23 in the text

# First let's create some data, namely a 2-D Gaussian dist
# of 1000 points with mean of in each dimension with a 
# stdev=2 in the first and stdev=1 in the second, all 
# rotated 45 degrees about the mean
import numpy as np

N = 1000
sigma1 = 2.0
sigma2 = 1.0
mu = [10, 10]
alpha_deg = 45.0
alpha = alpha_deg * np.pi / 180.

# Now import the "data-creator" from astroML
from astroML.stats.random import bivariate_normal

# And generate the data points
np.random.seed(0)
X = bivariate_normal(mu, sigma1, sigma2, alpha, N)

# Add some noise to the distribution.  Make 15% of the points
# share the same characteristics of the already made data, just
# make sigma1 = 2 and sigma2 = 4, rather than [2,1]. Do note that the
# figure caption in the text says sigma1 = sigma2 = 5. That's a lie.
# If you want, change the 2.0 and 4.0 both to 5.0 and see how the
# plot changes. 2.0 and 4.0 is what matches the figure image.
X[:int(.15 * N)] = bivariate_normal((10,10), 2.0, 4.0, alpha,
                                    int(.15 * N))

# Transpose the (1000,2) array to a (2,1000) array
# and unpack it into x,y
x, y = X.T

# Use astroML's fitting function to draw out statistics
from astroML.stats import fit_bivariate_normal

# Get the 'non-robust' fit statistics. These rely on finding the principal
# axes by estimating alpha with eqn. 3.92, where the Pearson r coefficient
# is from eqn. 3.102 and sigmas from eqn. 3.36.
#  via astroML documentation, 'non-robust' means using mean/stdev statistic
(mu_nr, sigma1_nr, sigma2_nr, alpha_nr) = fit_bivariate_normal(x, y, robust=False)

# And the 'robust' fit statistics. These use eqn. 3.92 again, but sub
# in eqn. 3.93 for r (ie: use eqn. 3.93 rather than 3.102) and replace
# the variance V values with sigma^2 from eqn. 3.36 (I think...)
# via astroML documentation, 'robust' -> rank-based statistics
(mu_r, sigma1_r, sigma2_r, alpha_r) = fit_bivariate_normal(x, y, robust=True)

# Create figure and axes for plotting purposes
fig = plt.figure()
ax = fig.add_subplot(111)

# And put the data points all up on it
ax.scatter(x, y, s=2, lw=0, c='k', alpha=0.5)

# Put some ellipses on the plot at 1 and 3 sigma
from matplotlib.patches import Ellipse
for Nsig in [1,3]:
    
    # True fit
    E = Ellipse(mu, sigma1 * Nsig, sigma2 * Nsig, alpha_deg,
                ec='k', fc='none')
    ax.add_patch(E)
    
    # Non-robust
    E = Ellipse(mu_nr, sigma1_nr * Nsig, sigma2_nr * Nsig, 
                alpha_nr * 180./np.pi, ec='k', fc='none',
                linestyle='dotted')
    ax.add_patch(E)
    
    # Robust
    E = Ellipse(mu_r, sigma1_r * Nsig, sigma2_r * Nsig,
                alpha_r * 180./np.pi, ec='k', fc='none',
                linestyle='dashed')
    ax.add_patch(E)
    
# Set plot limits
ax.set_xlim(5.5, 14.5)
ax.set_ylim(5.5, 14.5)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')

# Put together the authors' (by own admission) hacky method for making a 
# tolerable looking legend by plotting some fake data off the grid
ax.plot([0], [0], '-k', label='Input')
ax.plot([0], [0], ':k', label='Fit')
ax.plot([0], [0], '--k', label='Robust Fit')
ax.legend(loc='lower right')

plt.show()

    
```
